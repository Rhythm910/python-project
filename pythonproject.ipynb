```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#drop unnamed column
df= df.drop ("Unnamed: 0" , axis=1)
print(df.head())
```

```python
#change weekly study hours column
df["WklyStudyHours"] = df["WklyStudyHours"] .str.replace("05-oct","5-10")
print(df.head())
```

```python
#gender distribution
plt.figure(figsize=(5,5))
ax=sns.countplot(data= df, x= "Gender")
ax.bar_label(ax.containers[0])

plt.show()
```

#from above chart we have analysed that:
#the number of females is more in data than the number of males


```python
gb= df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot=True)
plt.figure(figsize=(4,4))
plt.show()
```

#from the above chart we have concluded that the education of the parents have a good impact on their scores

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot=True)
plt.title("Relationship between Parent's marital status and Student's score")
plt.figure(figsize=(4,4))
plt.show()
```

```python
#from the above chart we have concluded that there is no impact on student's scores due to their parent's marital status
```
